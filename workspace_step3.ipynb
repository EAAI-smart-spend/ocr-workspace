## 3.4 Train the model

````python

!CUDA_VISIBLE_DEVICES='0,1'

!python deep-text-recognition-benchmark/train.py \
    --train_data "dataset/step3/training" \
    --valid_data "dataset/step3/validation" \
    --select_data / \
    --batch_ratio 1 \
    --Transformation "TPS" \
    --FeatureExtraction "VGG" \
    --SequenceModeling "BiLSTM" \
    --Prediction "CTC" \
    --input_channel 1 \
    --output_channel 512 \
    --hidden_size 512 \
    --saved_model "pre_trained_models/chinese.pth" \
    --FT \
    --workers 0
````
